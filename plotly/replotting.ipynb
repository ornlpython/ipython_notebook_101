```python
from plotly.offline import plot, init_notebook_mode, iplot
init_notebook_mode()
import plotly.plotly as py
import plotly
import plotly.graph_objs as go
import numpy as np
```

### 1 plot at a time 

```python
N = 500
random_x = np.linspace(0, 1, N)
random_y = np.random.randn(N)

# Create a trace
trace = go.Scatter(
    x = random_x,
    y = random_y
)

data = [trace]

# Plot and embed in ipython notebook!
iplot(data)
```
