```python
from IPython.html import widgets
from IPython.display import display
```

### Example 1

```python
radio1 = widgets.RadioButtons(
    description='Color',
    options=['red', 'blue', 'green'])
display(radio1)
```

### Example 2

```python
radio2 = widgets.RadioButtons(
    options = {"red":'r', "blue":'b', "green":'g'})
display(radio2)
```

```python
print(radio2.value)
```
