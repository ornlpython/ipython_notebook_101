This is using the PIL library

```python
from PIL import Image
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
filename = 'images/Turb0096.tif'
```

# Loading a tif file (using PIL)

```python
im_pil = Image.open(filename)
print(type(im_pil))
```

```python
plt.figure(figsize=(10,10))
plt.imshow(im_pil)
plt.colorbar()
```

# Loading a tif file (using matplotlib)

```python
import matplotlib.image as mpimg
```

```python
im_matplotlib = mpimg.imread(filename)
print(type(im_matplotlib))
```

```python
plt.figure(figsize=(10,10))
plt.imshow(im_matplotlib)
plt.colorbar()
```
