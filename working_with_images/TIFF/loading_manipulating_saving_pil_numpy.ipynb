# Learning how to load (PIL), manipulate data (numpy)  and save them back in tiff (16bits) (PIL)

Python Imaging Library (PIL) documentation can be found [here](http://effbot.org/imagingbook/pil-index.htm)

```python
from PIL import Image
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
filename = 'images/Turb0096.tif'
```

### Loading file

```python
im = Image.open(filename)
```

```python
print("im.info: ", im.info)
print("im.format: ", im.format)
print("im.mode: ", im.mode)
print('im.format_description: ', im.format_description)
```

```python
im.getdata()
```

### Saving file

```python
output_filename = 'images/Turb0096_PIL.tif'
```

```python
im.save(output_filename)
```

### Loading again to make sure format is identical

```python
im_reloaded = Image.open(output_filename)
```

```python
print("im_reloaded.info: ", im_reloaded.info)
print("im_reloaded.format: ", im_reloaded.format)
print("im_reloaded.mode: ", im_reloaded.mode)
print('im_reloaded.format_description: ', im_reloaded.format_description)
```

# Perfect so far!!!
