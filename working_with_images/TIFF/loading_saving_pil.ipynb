This is using the PIL library

```python
from PIL import Image
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
filename = 'images/Turb0096.tif'
```

# Loading a tif file

```python
im = Image.open(filename)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(im)
plt.colorbar()
```

# Saving a tif file

```python
output_filename = 'images/Turb0096_modified.tif'
```

```python
im.save(output_filename)
```

# Checking that the initial image and the load/save/load image are the same

```python
im_original = Image.open(filename)
im_saved = Image.open(output_filename)
```

```python
plt.figure(figsize=(10,20))
ax_img1 = plt.subplot(1,2,1)
ax_img1.imshow(im_original)
ax_img1.set_title('original image')

ax_img2 = plt.subplot(1,2,2)
ax_img2.imshow(im_saved)
ax_img2.set_title('load/save/load image')

print(type(im_saved))
```

 The file will be saved as it was loaded (same format) 16bits [0-> 65535]
